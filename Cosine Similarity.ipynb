```python
import pandas as pd
```

```python
path="D:/H&M dataset/articles.csv"
df=pd.read_csv(path)
df.head()
```

```python
df=df[:100]
```

```python
def important_features(dataset):
    data=dataset.copy()
    for i in range(0, dataset.shape[0]):
        data["imp"]=data["prod_name"]+' '+data["product_type_name"]+" "+data["product_group_name"]
    return data
```

```python
data=important_features(df)
data.head()
```

```python
data["ids"]=[i for i in range(0,data.shape[0])]
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
vec=TfidfVectorizer()
```

```python
vecs=vec.fit_transform(data["imp"].apply(lambda x: np.str_(x)))
```

```python
vecs.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vecs) 
```

```python
sim[0][72]
```

```python
def recommend(title):
    product_id=data[data.prod_name==title]["ids"].values[0]
    scores=list(enumerate(sim[product_id]))
    sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
    sorted_scores=sorted_scores[1:]
    products=[data[products[0]==data["ids"]]["prod_name"].values[0] for products in sorted_scores]
    return products
```

```python
def recommend_ten(prod_list):
    first_ten=[]
    count=0
    for prod in prod_list:
        if count > 9:
            break
        count+=1
        first_ten.append(prod)
    return first_ten
```

```python
lst=recommend('Strap top')
m=recommend_ten(lst)
m
```

```python
df=df[:100]
df=df[["customer_id","article_id"]]
df
```

```python
path2="D:/H&M dataset/articles.csv"
df2=pd.read_csv(path2)
df2.head()
```

```python
df2=df2[["article_id","prod_name","product_type_name","product_group_name"]]
df2.head()
```

```python
result = pd.merge(df,df2, on="article_id")
result
```

```python
def important_features(dataset):
    data=dataset.copy()
    for i in range(0, dataset.shape[0]):
        data["imp"]=data["prod_name"]+' '+data["product_type_name"]+" "+data["product_group_name"]
    return data
```

```python
data=important_features(result)
data.head()
```

```python
data["ids"]=[i for i in range(0,data.shape[0])]
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
vec=TfidfVectorizer()
```

```python
vecs=vec.fit_transform(data["imp"].apply(lambda x: np.str_(x)))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vecs) 
```

```python
sim[0][0]
```

```python
def recommend(title):
    product_id=data[data.prod_name==title]["ids"].values[0]
    scores=list(enumerate(sim[product_id]))
    sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
    sorted_scores=sorted_scores[1:]
    products=[data[products[0]==data["ids"]]["article_id"].values[0] for products in sorted_scores]
    return products
```

```python
def recommend_ten(prod_list):
    first_ten=[]
    count=0
    for prod in prod_list:
        if count > 9:
            break
        count+=1
        first_ten.append(prod)
    return first_ten
```

```python
column_names = ["customer_id","prediction"]

answer = pd.DataFrame(columns = column_names)

answer.head()
```

```python
answer['customer_id']=data['customer_id']
answer.head()
```

```python
for i in range(26):
    products=recommend(data.prod_name[i])
    answer['prediction'][i]=recommend_ten(products)
    
```

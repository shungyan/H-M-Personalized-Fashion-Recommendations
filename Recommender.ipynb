```python
data = pd.melt(df.set_index('customer_id')['article_id'].apply(pd.Series).reset_index(), 
             id_vars=['customer_id'],
             value_name='article_id') \
    .dropna().drop(['variable'], axis=1) \
    .groupby(['customer_id', 'article_id']) \
    .agg({'article_id': 'count'}) \
    .rename(columns={'article_id': 'purchase_count'}) \
    .reset_index() \
    .rename(columns={'article_id': 'product_id'})

data['product_id'] = data['product_id'].astype(np.int64)
```

```python
data.head()
```

```python
def create_data_dummy(data):
    data_dummy = data.copy()
    data_dummy['purchase_dummy'] = 1
    return data_dummy
data_dummy = create_data_dummy(data)
```

```python
df_matrix = pd.pivot_table(data, values='purchase_count', index='customer_id', columns='product_id')
```

```python
df_matrix_norm = (df_matrix-df_matrix.min())/(df_matrix.max()-df_matrix.min())
```

```python
df_matrix_norm.head()
```

```python
d = df_matrix_norm.reset_index() 
d.index.names = ['scaled_purchase_freq'] 
data_norm = pd.melt(d, id_vars=['customer_id'], value_name='scaled_purchase_freq').dropna()
print(data_norm.shape)
data_norm.head()
```
